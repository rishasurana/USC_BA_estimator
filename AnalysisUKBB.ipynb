Concatenate all CSV files into one large data frame/CSV.

```python
import pandas as pd
import numpy as np
import os 
from glob import glob
```

```python
path_of_BA_csvs = '/Users/rishasurana/Documents/USC_BA_estimator/BACA/UKBBBA'
files = glob(os.path.join(path_of_BA_csvs, '*.csv'))
UKBB_BA_csv = pd.DataFrame()

UKBB_BA_csv = pd.concat([pd.read_csv(file) for file in files ], ignore_index=True)
UKBB_BA_csv.pop(UKBB_BA_csv.columns[0])

UKBB_BA_csv = UKBB_BA_csv.to_numpy()

UKBB_BA_csv
```

Make data frame of CA CSV.

```python
UKBB_CA_csv = pd.read_csv('/Users/rishasurana/Documents/USC_BA_estimator/BACA/UKBBCA.csv').to_numpy()
UKBB_CA_csv
```

Then find corresponding CAs and add that to data frame.

```python
CAList = []
for ID in UKBB_BA_csv:
    print(ID)
    index = np.where(UKBB_CA_csv == ID)[0]
    CA = UKBB_CA_csv[index, 1][0]
    CAList.append(CA)

mainArray = np.column_stack((UKBB_BA_csv, np.asanyarray(CAList)))
mainDF = pd.DataFrame(mainArray)
mainDF.columns = ['ID', 'BA', 'CA']
mainDF
```

Then do analysis on the final CSV file data. First, begin with subtracting BA and CA values. Then save as a final CSV file. Update as you run more subjects.

```python
diffList = []
for ID, BA, CA in mainArray:
    diffList.append(BA-CA)

mainDF.insert(3, 'DIFF', diffList, False)
mainDF.to_csv('tableUKBB.csv')
mainDF
```
